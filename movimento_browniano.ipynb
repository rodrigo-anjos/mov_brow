```python
from random import randrange
from matplotlib import pyplot

x=0
y=0
i=0
xlist=[0]
ylist=[0]
while i<1000000:
    z=randrange(4)
    if z==0:
        x=x+1
    if z==1:
        x=x-1
    if z==2:
        y=y+1
    if z==3:
        y=y-1
    xlist.append(x)
    ylist.append(y)
    i=i+1


    
pyplot.plot(xlist,ylist)
pyplot.show()


```
